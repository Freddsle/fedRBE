## ENTREZ ids

```python
id_lists = dict()

# Read the JSON file into a Python dictionary
with open('preprocessing_info/GBacc_IDs.json', 'r') as file:
    center_id_lists = json.load(file)

# transform list into pandas df column
for center in center_id_lists.keys():
    # use proteins as a colname of a dataframe
    id_lists[center] = pd.DataFrame(center_id_lists[center], columns=['GB_ACC'])
        
```

```python
upset_plot.generate_upset_plot(
    id_lists,
    colname='GB_ACC',
    title='Upset plot of ovarian cancer microarray data, GB ids',
    categories=list(id_lists.keys()),
    save_plot=True,
    path_svg='preprocessing_info/upset_plot_GBACC.svg',
)
```
